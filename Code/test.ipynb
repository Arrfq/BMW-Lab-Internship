```python
import pandas as pd
```

```python
ues_data_path = 'ues-2024-07-16-083100-qfnriic8.csv'
cells_data_path = 'cells-2024-07-16-083043-teg2vaqy.csv'

ues_data = pd.read_csv(ues_data_path)
cells_data = pd.read_csv(cells_data_path)
```

```python
ues_data.head()
ues_data['Viavi.UE.Rsrp']
```

```python
cells_data.head()
```

```python
ssb_data = ues_data[ues_data['Viavi.mMIMO.Type'] == 'SSB']
csirs_data = ues_data[ues_data['Viavi.mMIMO.Type'] == 'CSI-RS']
```

```python
ssb_ue_throughput_dl_mean = ssb_data['DRB.UEThpDl'].mean()
csirs_ue_throughput_dl_mean = csirs_data['DRB.UEThpDl'].mean()

ssb_ue_throughput_ul_mean = ssb_data['DRB.UEThpUl'].mean()
csirs_ue_throughput_ul_mean = csirs_data['DRB.UEThpUl'].mean()
```

```python
ssb_cells = ['S1', 'S2']
csirs_cells = ['S3', 'S4']
ssb_cell_data = cells_data[cells_data['Viavi.Cell.Name'].str.startswith(tuple(ssb_cells))]
csirs_cell_data = cells_data[cells_data['Viavi.Cell.Name'].str.startswith(tuple(csirs_cells))]
```

```python

ssb_ue_rsrp_mean = ssb_data['Viavi.UE.Rsrp'].mean()
csirs_ue_rsrp_mean = csirs_data['Viavi.UE.Rsrp'].mean()

ssb_cell_throughput_dl_mean = ssb_cell_data['DRB.UEThpDl'].mean()
csirs_cell_throughput_dl_mean = csirs_cell_data['DRB.UEThpDl'].mean()

ssb_cell_throughput_ul_mean = ssb_cell_data['DRB.UEThpUl'].mean()
csirs_cell_throughput_ul_mean = csirs_cell_data['DRB.UEThpUl'].mean()

throughput_rsrp_comparison_Cell = pd.DataFrame({
    'Metric': ['Throughput DL (Mbps)', 'Throughput UL (Mbps)'],
    'SSB (Cell)': [ssb_cell_throughput_dl_mean, ssb_cell_throughput_ul_mean],
    'CSI-RS (Cell)': [csirs_cell_throughput_dl_mean, csirs_cell_throughput_ul_mean]
})

throughput_rsrp_comparison_UE = pd.DataFrame({
    'Metric': ['Throughput DL (Mbps)', 'Throughput UL (Mbps)', 'RSRP (dBm)'],
    'SSB (UE)': [ssb_ue_throughput_dl_mean, ssb_ue_throughput_ul_mean, ssb_ue_rsrp_mean],
'CSI-RS (UE)': [csirs_ue_throughput_dl_mean, csirs_ue_throughput_ul_mean, csirs_ue_rsrp_mean]
})
```
