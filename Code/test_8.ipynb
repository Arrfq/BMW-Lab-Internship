```python
import json
import pandas as pd
import numpy as np
```

```python
ues_csirs_path = 'ue_csirs.csv'
cells_csirs_path = 'cells_csirs.csv'


ues_csirs = pd.read_csv(ues_csirs_path)
cells_csirs = pd.read_csv(cells_csirs_path)
```

```python
ues_ssb_path = 'ue_ssb.csv'
cells_ssb_path = 'cells_ssb.csv'

ues_ssb = pd.read_csv(ues_ssb_path)
cells_ssb = pd.read_csv(cells_ssb_path)
```

```python
ues_pedestrian_csirs = ues_csirs[ues_csirs['Viavi.UE.Name'].str.contains('Pedestrian')]
ues_pedestrian_ssb = ues_ssb[ues_ssb['Viavi.UE.Name'].str.contains('Pedestrian')]

ues_indoor_csirs = ues_csirs[ues_csirs['Viavi.UE.Name'].str.contains('Indoor')]
ues_indoor_ssb = ues_ssb[ues_ssb['Viavi.UE.Name'].str.contains('Indoor')]
```

```python
ue_pedestrian_csirs_throughput_dl_mean = ues_pedestrian_csirs['DRB.UEThpDl'].mean()
ue_pedestrian_csirs_throughput_ul_mean = ues_pedestrian_csirs['DRB.UEThpUl'].mean()
ue_pedestrian_csirs_rsrp_mean = ues_pedestrian_csirs['Viavi.UE.Rsrp'].mean()
```

```python
ue_pedestrian_ssb_throughput_dl_mean = ues_pedestrian_ssb['DRB.UEThpDl'].mean()
ue_pedestrian_ssb_throughput_ul_mean = ues_pedestrian_ssb['DRB.UEThpUl'].mean()
ue_pedestrian_ssb_rsrp_mean = ues_pedestrian_ssb['Viavi.UE.Rsrp'].mean()
```

```python
ue_indoor_csirs_throughput_dl_mean = ues_indoor_csirs['DRB.UEThpDl'].mean()
ue_indoor_csirs_throughput_ul_mean = ues_indoor_csirs['DRB.UEThpUl'].mean()
ue_indoor_csirs_rsrp_mean = ues_indoor_csirs['Viavi.UE.Rsrp'].mean()
```

```python
ue_indoor_ssb_throughput_dl_mean = ues_indoor_ssb['DRB.UEThpDl'].mean()
ue_indoor_ssb_throughput_ul_mean = ues_indoor_ssb['DRB.UEThpUl'].mean()
ue_indoor_ssb_rsrp_mean = ues_indoor_ssb['Viavi.UE.Rsrp'].mean()
```
